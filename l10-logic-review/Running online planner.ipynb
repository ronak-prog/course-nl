```python
# Based on http://solver.planning.domains/
import requests
```

```python
# domains from - https://github.com/AI-Planning/classical-domains/tree/master/classical/assembly
# stored locally
domainfile = "data/domain.pddl"
problemfile = "data/prob01.pddl" 
outputfile = "data/output.txt"
```

```python
import requests, sys

data = {'domain': open(domainfile, 'r').read(),
        'problem': open(problemfile, 'r').read()}

resp = requests.post('http://solver.planning.domains/solve',
                     verify=False, json=data).json()

```

```python
print (str(resp))
```

```python
if(resp['status'] == 'error'):
    print(resp['result']['output'])
else:
    with open(outputfile, 'w') as f:
        f.write('\n'.join([act['name'] for act in resp['result']['plan']]))
```
